```python
import pandas as pd
df = pd.read_csv("openpowerlifting.csv")
print(df.head())
```

```python
df = df[df['Tested'] == 'Yes']
df = df[df['TotalKg'] > 0]
df = df[df['Equipment'] == 'Raw']
df["Sex"] = df["Sex"].replace({"F": 1, "M": 2})
```

```python
df = df[['BodyweightKg', 'TotalKg', 'Sex', 'WeightClassKg',
         'Best3SquatKg','Best3BenchKg','Best3DeadliftKg', 'Age']]

df = df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cor = df[['BodyweightKg','Sex', 'Age','Best3SquatKg','Best3BenchKg','Best3DeadliftKg']].corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cor, annot=True)
plt.title('Correlation between lifting variables')
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(data=df, x='BodyweightKg', y='TotalKg',hue = 'WeightClassKg', palette = 'viridis', alpha = .6)
plt.show()
```
